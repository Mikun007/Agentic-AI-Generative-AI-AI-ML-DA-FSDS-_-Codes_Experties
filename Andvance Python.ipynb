## different type of logical statement in python

* if
* else
* nested if
* if elif else

```python
if True:
    print("Data Science")
```

```python
if False:
    print("Data Science")
print("bye for now")
```

```python
if True:
    print("Data Science")
print("Bye for now")
```

```python
if True:
    print("Data Science")
else:
    print("Bye for now")
```

```python
# Print a number is even number or odd number
number = int(input("What is the number you want to know: "))

if number%2 == 0:
    print(f"{number} is even number.")
else:
    print(f"{number} is odd number.")
```

```python
number = int(input("What is the number you want to know: "))

if number%2 != 0:
    print(f"{number} is odd number.")
else:
    print(f"{number} is even number.")
```
