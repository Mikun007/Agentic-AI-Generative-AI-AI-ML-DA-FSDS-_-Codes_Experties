# Loops in python

### while loop

```python
i = 1
while i<=5:
    print('data science');
    i+=1
```

```python
i =5 
while i>=1:
    print("Data science")
    i -= 1
```

```python
i = 5

while i>=1:
    print("data science: ", i)
    i -=1
```

```python
i = 1

while i<=5:
    print("data science")
    j = 1
    while j<=4:
        print("technology")
        j += 1

    i += 1
    print()
```

```python
i = 1

while i<=5:
    print("data science:", end=" ")
    j = 1
    while j<=4:
        print("technology", end=" ")
        j += 1

    i += 1
    print()
```

```python
i = 1
while i<=2:
    j = 0
    while j<=2:
        print(i*j, end=" ")
        j+=1
    print()
    i+=1
    
    
```

```python
i = 1
while i<=4:
    j = 0
    while j<=3:
        print(i*j, end=" ")
        j+=1
    print()
    i+=1
```

```python
# while else
i = 10

while i <=4:
    j = 1
    while j<=3:
        print(i*j, end=" ")
        j +=1
    print()
    i +=1
else:
    print("condition is not match")
```

```python
x = int(input("How many chocolets you want: ?"))
i = 1
while i<=x:
    print("Chocolet :", i)
    i+=1
```

```python
available_chocolet = 5 # The machine has only 5 chocolet
x = int(input("How may chocolets you want:?"))

i = 1

while i<=x:
    if i>available_chocolet:
        print("out of stock")
        break
    print("Chocolet: ", i)
    i+=1

print("Bye for now")
```

#### Continue

```python
for i in range(1, 31):
    if i%3 == 0:
        continue # Skipping all the 3 table
    print(i)
print('end') 
```

```python
for i in range(1,51):
    if i%3 ==0 or i%5==0:
        continue #skipping all the 3 and 5 table
    print(i)
```
