# Loops

## for loop

### Loops with list

```python
li = ["geeks", "for", "geeks"]

for x in li:
    print(x)
```

### Loops with tuple

```python
tup = ("geeks", "for", "geeks")
for x in  tup:
    print(x)
```

### Loops with string

```python
s = "abc"
for char in s:
    print(char)
```

### Loops with dictionary

```python
d = {"x":123,
     "y": 354
    }
for x in d:
    print(f"{x}: {d[x]}")
```

### Loops through sets

```python
set1 = {10, 30, 20}
for x in set1:
    print(x)
```

## while loop

```python
cnt = 0
while (cnt < 3):
    cnt = cnt + 1
    print("Hello Geek")
```

#### Nested loops

```python
for i in range(1,5):
    for j in range(i):
        print(i, end = " ")
    print()
```

### Loop Control Statements

```python
for letter in "geeksforgeeks":
    if letter == "e" or letter =="s":
        continue
    print("Current letter: ", letter)
```

```python
for letter in "GeeksForGeeks":
    if letter == "e" or letter == "s":
        break

print("Current Letter: ", letter)
```

# Loop Problems

## For Loop

```python
# You are given a number N, you need to print its multiplication table

num = int(input("Give a number to have multiplication table"))

for i in range(1, 11):
    print(i*num)
```

```python
"""
    You are given a number n, take input of n and print its
     multiplication table in a single line using for loop till n*10.
"""
n = 5

for i in range(1, 11):
    print(i*n, end=" ")
```

```python
""" 
    You are given a string s, you need to print its
        characters at even indices (index start at 0).
"""
s = "DoctorPhenomenal"

for i in range(0, len(s), 2):
    print(s[i], end="")
```

```python
# Other way
s = "DoctorPhenomenal"

for i in range(0, len(s)):
    if i%2 == 0:
        print(s[i], end="")
```

## While Loop

```python
"""
    Given a number x, the task is to print the numbers
        from x to 0 in decreasing order in a single line.
"""
x = 5

while x>=0:
    print(x, end=" ")
    x -= 1
```

```python
"""
Given a positive integer x, teh task is to print the number from 1 to x in the order as
1^2, 2^2, 3^2,... (in increasing order)
"""

x = 10
i = 1
squared = 0
while True:
    if squared > x:
        break
    squared = i**2
    print(squared, end=" ")
    i+=1
    x-=1
```
