## 1. Right Angle Triangle Pattern

```python
for i in range(1, 6):
    print(" * " * i)
```

## 2. Inverted Right Angle Triangle Pattern

```python
for i in range(6, 0, -1):
    print(" * " * i)
```

## 3. Pyramid Pattern

```python
for i in range(1, 6):
    print(''*(5-i)+' * '*(2*i-1))
```

## 4. Inverted Pyramid Pattern

```python
for i in range(6, 0, -1):
    print(" * "*(2*i-1))
```

## 5. Diamond Pattern

```python
for i in range(1, 6):
    print(" * "*(2*i-1))
for i in range(6, 0, -1):
    print(" * "*(2*i-1))
```

## 6. Hallow Square Pattern

```python
for i in range(5):
    for j in range(5):
        if i==0 or i==4 or j==0 or j==4:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```

## 7. Full Square Pattern

```python
for i in range(6):
    print(" * "*6)
```

## 8. Right Angle triangle (Number Pattern)

```python
for i in range(1, 6):
    print(" ".join(str(x) for x in range(1, i+1)))
```

## 9. Inverted Right Angle Triangle (number pattern)

```python
for i in range(6, 0, -1):
    print(" ".join(str(x) for x in range(1, i+1)))
```

## 10. floyd's triangle

```python
num =1
for i in range(1, 6):
    for j in range(0, i):
        print(num, end=' ')
        num += 1
    print()
```

## 11. Hallow Right Angle Triangle

```python
for i in range(1, 6):
    for j in range(1, i+1):
        if j == 1 or j==i or i==5:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

## 12. Hallow Pyramid Patter

```python
for i in range(1,6):
    for j in range(5-i):
        print(" ", end=" ")
    for j in range(2*i-1):
        if j == 0 or j == 2*i-2 or i ==5:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

## 13. Hallow Diamond Pattern

```python
n = 5
for i in range(1, n+1):
    for j in range(n -i):
        print(" ", end=" ")
    for j in range(2*i-1):
        if j ==0 or j==2*i-2:
            print("*", end = " ")
        else:
            print(" ", end=" ")
    print()

for i in range(n-1, 0, -1):
    for j in range(n-i):
        print(" ", end= " ")
    for j in range(2*i-1):
        if j == 0 or j == 2*i-2:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

## 14. Hallow Diamond (Number Pattern)

```python
n = 5
for i in range(1, n+1):
    for j in range(n-i):
        print(" ", end=" ")
    for j in range(2*i-1):
        if j == 0 or j ==2*i-2:
            print(i, end= " ")
        else:
            print(" ", end=" ")
    print()

for i in range(n-1, 0, -1):
    for j in range(n-i):
        print(" ", end=" ")
    for j in range(2*i-1):
        if j==0 or j==2*i-2:
            print(i, end=" ")
        else:
            print(" ", end=" ")
    print()
```

## 15. Butterfly Pattern

```python
n = 5
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end=" ")
    for j in range(2*(n-i)):
        print(" ", end=" ")
    for j in range(1, i+1):
        print(j, end=" ")
    print()

for i in range(n, 0, -1):
    for j in range(1, i+1):
        print(j, end=" ")
    for j in range(2*(n-i)):
        print(" ", end=" ")
    for j in range(1, i+1):
        print(j, end=" ")
    print()

n = 5
for i in range(1, n+1):
    for j in range(i):
        print("*", end=" ")
    for j in range(2*(n-i)):
        print(" ", end=" ")
    for j in range(i):
        print("*", end=" ")
    print()

for i in range(n, 0, -1):
    for j in range(i):
        print("*", end=" ")
    for j in range(2*(n-i)):
        print(" ", end=" ")
    for j in range(i):
        print("*", end=" ")
    print()
```

## 16. Hallow Number Pyramid

```python
n= 5
for i in range(1, n+1):
    for j in range(n-i):
        print(" ", end=" ")
    for j in range(1, 2*i):
        if j==1 or j==2*i-1 or i==n:
            print(i, end=" ")
        else:
            print(" ", end=" ")
    print()
```

## 17. Full Star Pyramid

```python
n = 5

for i in range(1, n+1):
    for j in range(n-i):
        print(" ", end=" ")

    for j in range(2*i-1):
        print("*", end=" ")

    print()
```

## 18. Inverted full Star Pyramid

```python
n = 5

for i in range(n, 0, -1):
    for j in range(n-i):
        print(" ", end=" ")

    for j in range(2*i-1):
        print("*", end=" ")
    print()
```

## 19. Left Aligned Pyramind Pattern

```python
n = 5

for i in range(1, n+1):
    for j in range(i):
        print("*", end=" ")
    print()

n = 5

for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

## 20. Right Aligned Pyramid Pattern

```python
n = 5

for i in range(1, n+1):
    for j in range(n-i):
        print(" ", end=" ")

    for j in range(1, i+1):
        print(j, end=" ")

    print()

n = 5

for i in range(1, n+1):
    for j in range(n-i):
        print(" ", end=" ")
    for j in range(1, i+1):
        print("*", end=" ")
```
