```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

```python
# Create an image data generator with augmentations

datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.4,
    height_shift_range=0.4,
    shear_range=0.4,
    zoom_range=0.4,
    horizontal_flip=True,
    rescale=None,
    fill_mode="reflect" # nearest, reflect, wrap, constant
)

# load image from path
img =  load_img(r"C:\Users\mikun\Downloads\Gemini_Generated_Image_g5xljlg5xljlg5xl.png")
```

```python
img
```

```python
# convert to numpy array
x = img_to_array(img)
x = x.reshape((1,) + x.shape) # Reshape to (1, height, width, channels)

# Generate augmented images and save
i = 0
for batch in datagen.flow(x, batch_size = 1,
                          save_to_dir=r"D:\Mikun\All Learnings\Data Science PyCharm\CNN\different generated image",
                          save_prefix="sad", save_format="jpeg"):
    i += 1
    if i > 30:
        break # Prevent infinite loop
```
