```python
sent = "sam is natural when it comes to drawing"
sent_tokens = word_tokenize(sent)
sent_tokens
```

```python
for token in sent_tokens:
    print(nltk.pos_tag([token])) # difining all the grammer in the sentence
```

```python
sent2 = "John is eating a delicious cake"
sent2_tokens = word_tokenize(sent2)

for token in sent2_tokens:
    print(nltk.pos_tag([token]))
```

```python
from nltk import ne_chunk
```

```python
NE_Sent = "The US president stays in the WHITEHOUSE"
```

```python
NE_tokens = word_tokenize(NE_Sent) # tokenizing
# after tokenize need to add the pos tags
NE_tokens
```

```python
NE_tags = nltk.pos_tag(NE_tokens) # grammer conversion
NE_tags
```

```python
NE_NER = ne_chunk(NE_tags) # Entity recognition
print(NE_NER)
```
