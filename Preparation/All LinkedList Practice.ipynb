# LinkedList

### Singly LinkedList

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        temp.next = new_node

    def insert_at_beggining(self, data):
        new_node = Node(data)

        new_node.next = self.head
        self.head = new_node

    def insert_at_key(self, data, key):
        new_node = Node(data)
        temp = self.head

        if key == 0:
            print("Please call insert_at_beggining")
            return

        for _ in range(key-1):
            temp = temp.next

        new_node.next = temp.next
        temp.next = new_node

    def delete_node(self, data):
        if not self.head:
            print("There is nothing to delete")
            return

        temp = self.head
        while True:
            try:
                if temp.next.data == data:
                    break
            except:
                print("There is nothing like this node to delete")
                return
                
            temp = temp.next

        temp.next = temp.next.next
        
        

    def traversal(self):
        temp = self.head
        while temp:
            print(temp.data, end="->")
            temp = temp.next
        

root = LinkedList()
root.insert_at_end(4)
root.insert_at_end(3)
root.insert_at_end(2)
root.insert_at_end(1)

root.insert_at_beggining(5)
root.insert_at_beggining(6)

root.insert_at_key(7, 2)

root.delete_node(9)

root.traversal()
```

## <p>Doubly LinkedList</p>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beggining(self, data):
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            return

        temp = self.head

        temp.prev = new_node
        new_node.next = temp
        self.head = new_node

    def insertion_at_end(self, data):
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next

        new_node.prev = temp
        temp.next = new_node

    def insertion_at_key(self, data, key):
        new_node = Node(data)

        temp = self.head
        for _ in range(key-1):
            temp = temp.next

        new_node.prev = temp
        new_node.next = temp.next
        temp.next.prev = new_node
        temp.next = new_node

    def delete_node(self, data):
        if not self.head:
            print("There is nothing to delete")
            return

        temp = self.head
        while True:
            if temp.next.data == data:
                break
            temp = temp.next

        temp.next.next.prev = temp
        temp.next = temp.next.next

    def traversal(self):
        temp = self.head
        # prev = temp.data
        while temp:
            print(temp.data, end="<->")
            temp = temp.next

    def traversal_reverse(self):
        temp = self.head

        while temp.next:
            temp = temp.next
        while temp:
            print(temp.data, end="<->")
            temp = temp.prev

    
            

root = DoublyLinkedList()
root.insert_at_beggining(3)
root.insert_at_beggining(2)
root.insert_at_beggining(1)

root.insertion_at_end(7)
root.insertion_at_end(8)

root.insertion_at_key(4, 3)
root.insertion_at_key(5, 4)
root.insertion_at_key(6, 5)

root.delete_node(5)

root.traversal()
print()
root.traversal_reverse()
```

## <p>Circular LinkedList</p>

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class CircularLinkedList:
    def __init__(self):
        self.head = None

    def insertion_at_beggining(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return

        # print(data)
        temp = self.head
        while temp.next != self.head:
            temp = temp.next

        new_node.next = self.head
        temp.next = new_node
        self.head = new_node

    def insert_at_end(self, data):
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return

        curr = self.head
        while curr.next != self.head:
            curr = curr.next

        new_node.next = self.head
        curr.next = new_node

    def insert_at_key(self, data, key):
        new_node = Node(data)

        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return

        if key == 0:
            print("Please call insertion at beggining")
            return

        temp = self.head
        for _ in range(key-1):
            temp = temp.next

        new_node.next = temp.next
        temp.next = new_node

    def delete_node(self, data):
        temp = self.head
        while True:
            if temp.next.data == data:
                break

            temp = temp.next

        temp.next = temp.next.next
        
        

    def traversal(self):
        temp = self.head

        curr = temp.data
        while temp:
            print(temp.data, end="->")
            temp = temp.next
            if temp.data == curr:
                break

root = CircularLinkedList()
root.insertion_at_beggining(4)
root.insertion_at_beggining(3)
root.insertion_at_beggining(2)
root.insertion_at_beggining(1)

root.insert_at_end(5)
root.insert_at_end(6)
root.insert_at_end(7)

root.insert_at_key(9, 1)

root.delete_node(9)

root.traversal()
```
