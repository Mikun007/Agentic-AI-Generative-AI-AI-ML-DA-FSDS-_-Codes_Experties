```python
# josephus problem
def josephus(arr, k):
    index = 0
    while len(arr) > 1:
        index = (index+k-1) %len(arr)
        arr.pop(index)
    return arr[0]

N = 7
k = 3

lists = [i+1 for i in range(N)]

josephus(lists, k)
```

```python
# josephus another problem
def find_winner(n, k):
    survivor = 0
    for i in range(2, n + 1):
        survivor = (survivor + k) % i
    return survivor + 1 # Convert to 1-based index

print(f"The winner is: {find_winner(7, 3)}")
```

```python
def tower_of_hanoi(n, source, destination, auxiliary):
    # Base Case: If there's only one disk, move it directly
    if n == 1:
        print(f"Move disk 1 from {source} to {destination}")
        return
    
    # Step 1: Move n-1 disks from source to auxiliary rod
    tower_of_hanoi(n - 1, source, auxiliary, destination)
    
    # Step 2: Move the nth (largest) disk from source to destination rod
    print(f"Move disk {n} from {source} to {destination}")
    
    # Step 3: Move the n-1 disks from auxiliary rod to destination rod
    tower_of_hanoi(n - 1, auxiliary, destination, source)

# Example usage: 3 disks, starting at 'A', moving to 'C'
n = 3
tower_of_hanoi(n, 'A', 'C', 'B')
```

```python
def toh(n, s, d, a):
    if n == 1:
        print(f"move disk {n} from {s} to {d}")
        return
    toh(n-1, s, a, d)
    print(f"Move disk {n} from {s} to {d}")
    toh(n-1, a, d, s)

toh(3, "a", "b", "c")
```

```python
# Merge Sort
def merge_sort(arr):
    if len(arr) < 2:
        return arr

    mid = len(arr)//2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    sorted_array = []
    i, j = 0, 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_array.append(left[i])
            i += 1
        else:
            sorted_array.append(right[j])
            j += 1
    sorted_array.extend(left[i:])
    sorted_array.extend(right[j:])
    return sorted_array

arr = [9, 10, 3, 1, 5, 2]

merge_sort(arr)
```

```python
# longest substring finding

def longest_substring(s):
    seen = {}
    left = 0
    
    max_length = 0

    for right, char in enumerate(s):
        if char in seen and seen[char] >= left:
            left = seen[char] + 1
            
        seen[char] = right

        max_length = max(max_length, right - left + 1)

    return max_length

longest_substring("abcdabcabc")
```

```python
# Permutation Findings

def find_permutation(lst):
    if len(lst) <= 1:
        return [lst]
    result = []

    for i in range(len(lst)):
        current_element = lst[i]
        remaining_element = lst[:i] + lst[i+1:]

        for p in find_permutation(remaining_element):
            result.append([current_element] + p)
    return result

for i in find_permutation([1, 2, 3]):
    print(i)
```

```python
## Binary Tree

class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

def treeInsertion(root, data):
    if root is None:
        return Node(data)

    if data < root.data:
        root.left = treeInsertion(root.left, data)
    elif data > root.data:
        root.right = treeInsertion(root.right, data)
        
    return root

#BFS (Breadth First Search)
def BFS(root):
    if root:
        queue = [root]

        while queue:
            current = queue.pop(0)
            print(current.data, end="->")

            if current.left:
                queue.append(current.left)
            if current.right:
                queue.append(current.right)
                
# DFS (Depth First Search)
def inorder(root):
    if root:
        inorder(root.left)
        print(root.data, end="->")
        inorder(root.right)

def preOrder(root):
    if root:
        print(root.data, end="->")
        preOrder(root.left)
        preOrder(root.right)

def postOrder(root):
    if root:
        postOrder(root.left)
        postOrder(root.right)
        print(root.data, end='->')
        

node = Node(5)
treeInsertion(node, 2)
treeInsertion(node, 3)
treeInsertion(node, 4)
treeInsertion(node, 7)

print("This is BFS: ")
BFS(node)
print()

print()
print("This is DFS: ")
print("InOrder")
inorder(node)
print()

print("preOrder")
preOrder(node)
print()

print("postOrder")
postOrder(node)
print()
```

```python
#linked list

class Node():
    def __init__(self, data):
        self.data = data
        self.next = None
 
def insertBeginning(data, root):
    new_node = Node(data)

    new_node.next = root
    root = new_node
    return root

def insertEnd(data, root):
    new_node = Node(data)

    temp = root
    while temp.next:
        temp = temp.next

    temp.next = new_node
    return root

def insertAtKey(data, root, key):

    if key == 0:
        return insertBeginning(data, root)
    new_node = Node(data)

    temp = root
    for _ in range(key-1):
        temp = temp.next

    new_node.next = temp.next
    temp.next = new_node
    return root

def deleteData(data, root):
    if root:
        prev = None
        temp = root

        if data == temp.data:
            root = temp.next
            return root
            
        while temp.next:
            if temp.data == data:
                break
            else:
                prev = temp
                temp = temp.next

        prev.next = temp.next
        return root
        
    
def traversal(root):
    temp = root
    while temp:
        print(temp.data, end="->")
        temp = temp.next
        
ll = Node(1)
ll = insertBeginning(2, ll)
ll = insertBeginning(3, ll)
ll = insertEnd(4, ll)
ll = insertAtKey(5, ll, 2)
ll = deleteData(3, ll)
traversal(ll)
```

```python
# doubly linkedlist
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

def insertBeginning(data, root):
    newNode = Node(data)

    newNode.next = root
    root.prev = newNode
    root = newNode
    return root

def insert_at_end(data, root):
    new_node = Node(data)

    temp = root
    while temp.next:
        temp = temp.next

    temp.next = new_node
    new_node.prev = temp
    return root

def traversal(root):
    temp = root
    while temp:
        print(temp.data, end="<->")
        temp = temp.next

def traversal_u(root):
    temp = root
    while temp.next:
        temp = temp.next

    while temp:
        print(temp.data, end="<->")
        temp = temp.prev
        
dll = Node(1)

dll = insert_at_end(2, dll)
dll = insert_at_end(3, dll)
dll = insert_at_end(4, dll)

traversal(dll)
print()
traversal_u(dll)
```

```python
# Circular linkedlist
class Node():
    def __init__(self, data):
        self.data = data
        self.next = None

def insertion_at_end(root, data):
    new_node = Node(data)
    temp = root
    while temp.next != root:
        temp = temp.next
    temp.next = new_node
    new_node.next = root
    return root

def traversal(root):
    if root:
        temp = root
        while temp:
            print(temp.data, end="->")
            temp = temp.next
            if temp.data == root.data:
                break
cll = Node(1)
cll.next = cll
cll = insertion_at_end(cll, 2)
cll = insertion_at_end(cll, 3)
cll = insertion_at_end(cll, 4)
cll = insertion_at_end(cll, 5)
cll = insertion_at_end(cll, 6)

traversal(cll)
```

```python
# Graphs
graph = {
    'A': ['B', 'C', 'D'],
    'B': ['A', 'E', 'F'],
    'C': ['A'],
    'D': ['A', 'G'],
    'E': ['B'],
    'F': ['B'],
    'G': ['D']
}

# BFS
def bfs(graph, start):
    visited = set(start)
    queue = [start]

    while queue:
        node = queue.pop(0)
        print(node, end="->")

        for neighbour in graph[node]:
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(neighbour)

bfs(graph, "A")

# DFS
def dfs(graph, start, visited):
    visited.add(start)
    print(start, end="->")

    for neighbour in graph[start]:
        if neighbour not in visited:
            dfs(graph, neighbour, visited)

print()
visited = set()
dfs(graph, "A", visited)
```

| Operation         | Expression         | Purpose                      |
| ----------------- | ------------------ | ---------------------------- |

| Clear ith bit     | `n & ~(1 << i)`    | Make bit `i` equal to `0`    |
| Toggle ith bit    | `n ^ (1 << i)`     | Flip bit `i`                 |
| Even/Odd          | `n & 1`            | Check if number is odd       |
| Multiply by `2^k` | `n << k`           | Fast multiplication          |
| Divide by `2^k`   | `n >> k`           | Fast integer division        |
| Power of 2        | `(n & (n-1)) == 0` | Check if `n` is a power of 2 |
