## Recursion Problems

### `1. find the factorial of a number using recursion`

```python
def recursion(n=5):
    if n == 1:
        return n
    return n*recursion(n-1)

recursion(5)
```

### `2. Recursive function to generate the n-th fibonacci sequence`

```python
def fibonachi(n=9, f=0, s=1):
    if n <= 0:
        return
    print(s)
    return fibonachi(n = n-1, f= s, s=f+s)
    
fibonachi()
```

```python
def fibonacci_calculation(n=9):
    if n <=1:
        return n
    return fibonacci_calculation(n-1) + fibonacci_calculation(n-2)

fibonacci_calculation()
```

### `3. Sum of the first n natural number`

```python
def sum_of_nNumber(n=5):
    if n <=0:
        return n
    return n + sum_of_nNumber(n-1)
sum_of_nNumber(5)
```

### `4. Power of a number`

```python
def power_of_number(x=5, n=3):
    if n <= 0:
        return 1
    return x * power_of_number(x, n-1)

power_of_number()
```

### `5. Reverse a String`

```python
def reverse_string(s="Mikun"):
    if s == "":
        return ""
    return reverse_string(s[1:]) + s[:1]
reverse_string()
```

### `6. Sum of Digits of a number`

```python
def sum_of_digits(n=543):
    if n <= 0:
        return 0
    return (n%10) + sum_of_digits(n//10)

sum_of_digits()
```

### `7. Count the occurance of a character`

```python
def count_occurance(s="Mikuun", cal = {}):
    if s == "":
        return

    count_occurance(s[1:], cal)
    cal[s[0]] = cal.get(s[0], 0)+1
    return cal

count_occurance()
```

### `8. Greatest Common Divisor (GCD)`

```python
def gcd(a=48, b=18):
    if a == 0 or b==0:
        return a
    if a > b:
        calculation = b*(a//b)
        remaining = a - calculation
        return gcd(b, remaining)
    elif a < b:
        calculation = a*(b//a)
        remaining = b - calculation
        return gcd(a, remaining)

gcd(5, 8)
```

```python
def gcd(a, b):
    if b == 0:
        return a
    return gcd(b, a%b)
gcd(18, 48)
```
