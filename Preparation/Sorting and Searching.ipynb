### Bubble Sort

```python
arr = [5, 2, 1, 3, 0, 7, 4, 6]

for i in range(len(arr)):
    for j in range(0, len(arr)-i-1):
        if arr[i] < arr[j]:
            arr[i], arr[j] = arr[j], arr[i]

arr
```

## Selection Sort

```python
arr = [5, 2, 1, 3, 0, 7, 4, 6]# finding the min-index

for i in range(len(arr)):
    min_index = i
    for j in range(i, len(arr)):
        if arr[j] < arr[min_index]:
            min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]

arr
```

## Merge Sort

```python
arr = [5, 2, 1, 3, 0, 7, 4, 6] # Divide and Conquour

def merge_sort(arr):
    if len(arr) < 2:
        return arr

    mid = len(arr)//2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)

def merge(left, right):
    sorted_arr = []

    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_arr.append(left[i])
            i+=1
        else:
            sorted_arr.append(right[j])
            j+=1
    sorted_arr += left[i:]
    sorted_arr += right[j:]
    return sorted_arr

arr = [5, 2, 1, 3, 0, 7, 4, 6]
arr = merge_sort(arr)
arr
```

## Linear Search

```python
key = 5
for i in arr:
    if i == key:
        print(f"Found the key: {key} at index: {arr.index(key)}")
        break
```

## Binary Search

```python
left = 0
right = len(arr)-1
key = 7

while left <= right:
    mid = (left+right)//2

    if arr[mid] == key:
        print(f"found the key: {key} at index: {mid}")
        break

    if arr[mid] < key:
        left = mid+1

    if arr[mid] > key:
        right = mid-1
```
