```python
# find second largest

arr = [2, 4, 1, 6,4, 2, 4, 7, 9, 64]
largest, second_largest = 0, 0
for i in arr:
    if i > largest:
        largest, second_largest = i, largest
    elif i < largest and i > second_largest:
        second_largest = i

print(largest, second_largest)
```

```python
# Two Sum
arr = [11, 15, 2, 7, 9, 0, 5, 2]
target = 9

first, second = 0, 0
for i in range(len(arr)-1):
    first, second = arr[i], arr[i+1]

    if (first+second) == target:
        break
print(arr.index(first), arr.index(second))
```

```python
# maximum subarray finding
arr = [-1, 1, -3, 4, -1, 2, 1,-5, 4]

max_sum = arr[0]
curr_sum = arr[0]

for i in range(1, len(arr)):
    curr_sum = max(arr[i], arr[i]+curr_sum)
    max_sum = max(max_sum, curr_sum)

max_sum
```

```python
# move all zeroes to end
arr = [0, 1, 0, 3, 12]
new_arr = []
zero_counts = 0
for i in arr:
    if i == 0:
        zero_counts += 1
    else:
        new_arr.append(i)
new_arr.extend([0]*zero_counts)
new_arr
```

```python
def harami(arr, k):
    index = 0
    while len(arr) > 1:
        index = (index+k-1) %len(arr)
        arr.pop(index)
    return arr[0]

N = 7
k = 3

lists = [i+1 for i in range(N)]

harami(lists, k)
```

```python
# nth fibonachi series
n = 9
prev, now = 0, 1

for i in range(1, n):
    print(prev)
    prev, now = now , prev+now
```

```python
#climbing stairs
n = 5

first, second = 1, 2

for i in range(3, n+1):
    print(first, second)
    third = first + second
    first, second = second, third

print(first, second)
```
