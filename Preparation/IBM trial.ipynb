```python
# josephus problem
def harami(arr, k):
    index = 0
    while len(arr) > 1:
        index = (index+k-1) %len(arr)
        arr.pop(index)
    return arr[0]

N = 7
k = 3

lists = [i+1 for i in range(N)]

harami(lists, k)
```

```python
def tower_of_hanoi(n, source, destination, auxiliary):
    # Base Case: If there's only one disk, move it directly
    if n == 1:
        print(f"Move disk 1 from {source} to {destination}")
        return
    
    # Step 1: Move n-1 disks from source to auxiliary rod
    tower_of_hanoi(n - 1, source, auxiliary, destination)
    
    # Step 2: Move the nth (largest) disk from source to destination rod
    print(f"Move disk {n} from {source} to {destination}")
    
    # Step 3: Move the n-1 disks from auxiliary rod to destination rod
    tower_of_hanoi(n - 1, auxiliary, destination, source)

# Example usage: 3 disks, starting at 'A', moving to 'C'
n = 3
tower_of_hanoi(n, 'A', 'C', 'B')
```

### Sorting and searching

```python
# Bubble sort
arr = [2, 3, 61, 6, 4, 1, 9]

for i in range(len(arr)):
    for j in range(0, len(arr)-i-1):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]

arr
```

```python
# selection sort

arr = [2, 1, 5, 4, 2, 9, 62]

for i in range(len(arr)):
    min_index = i
    for j in range(i, len(arr)):
        if arr[j] < arr[min_index]:
            min_index = j
    arr[i], arr[min_index] = arr[min_index], arr[i]

arr
```

```python
def merge_sort(arr):
    # Base case: A list with 0 or 1 element is already sorted
    if len(arr) <= 1:
        return arr

    # 1. Divide: Find the midpoint and split the array
    mid = len(arr) // 2
    left_half = merge_sort(arr[:mid])
    right_half = merge_sort(arr[mid:])

    # 2. Conquer & Merge: Combine the sorted halves
    return merge(left_half, right_half)

def merge(left, right):
    sorted_arr = []
    i, j = 0, 0

    # Compare elements from both halves and add the smaller one
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            sorted_arr.append(left[i])
            i += 1
        else:
            sorted_arr.append(right[j])
            j += 1

    # Add any remaining elements from either half
    sorted_arr.extend(left[i:])
    sorted_arr.extend(right[j:])
    
    return sorted_arr

# Example usage:
my_list = [38, 27, 43, 3, 9, 82, 10]
sorted_list = merge_sort(my_list)
print(sorted_list) # Output: [3, 9, 10, 27, 38, 43, 82]
```

```python
# binary search
arr = [ 2, 3, 1, 5, 9, 32, 19]

for i in range(len(arr)):
    for j in range(0, len(arr)-i-1):
        if arr[j] > arr[j+1]:
            arr[j], arr[j+1] = arr[j+1], arr[j]

print(arr)

low = 0
high = len(arr) - 1

key = 9
while low <= high:
    mid = (low+high)//2

    if arr[mid] == key:
        print(mid)
        break
    elif arr[mid] < key:
        low = mid+1
    else:
        high = mid-1
else:
    print("Element now found")
```
