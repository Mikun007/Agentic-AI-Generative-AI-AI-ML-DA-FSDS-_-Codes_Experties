```python
# Q10.
string  = input("input a string: ")

reverse = string[::-1]

if string == reverse:
    print(f"The string is palindrome as string({string}) == reverse({reverse})")
else:
    print(f"The string is not palindrome as string({string}) != reverse({reverse})")
```
