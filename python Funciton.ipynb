```python
# can you pass a list of funciotn and we need to count from the list no. of even and odd number

def evenOdd(list):
    even = 0
    odd = 0
    for i in list:
        if i % 2 == 0:
            even += 1
        else:
            odd += 1
    print(even, odd)

evenOdd([2, 3, 4, 5, 6])
```

```python
# fibonachi sequence
def fibonachi(n):
    prev = 0
    now = 1

    print(prev)
    print(now)

    for i in range(n):
        temp = prev + now
        prev = now
        now = temp

        print(temp)

fibonachi(10)
```

Factorial

```python
def fact(n):
    if n==0:
        return 1
    return n * fact(n-1)

result = fact(6)
result
```

```python
n = 3

fact = 1
for i in range(1, n+1):
    fact *= i

print(fact)
```

## filter()

```python
def is_even(n):
    return n%2 == 0

nums = [3, 2, 6, 8, 4, 6, 2, 9]

evens = list(filter(is_even, nums))
print(evens)
```

```python
def is_odd(n):
    return n%2 != 0

nums = [3, 2, 6, 8, 4, 6, 2, 9]

odds = list(filter(is_odd, nums))
print(odds)
```

## map()

```python
def update(n):
    return n+2

nums = [3, 2, 6, 8, 4, 6, 2, 9]

evens = list(filter(is_even, nums))
double = list(map(update, evens))

print(evens)
print(double)
```

```python
nums = [3, 2, 6, 8, 4, 6, 2, 9]

evens = list(filter(is_even, nums))
double = list(map(lambda n: n*2, evens))
print(evens)
print(double)
```

## reduce()

```python
from functools import reduce

def add_all(a,b):
    return a+b

nums = [3, 2, 6, 8, 4, 6, 2]

evens = list(filter(is_even, nums))
double = list(map(lambda n: n*2, evens))

sums = reduce(add_all, double)

sums

print(sums)
```

```python
from functools import reduce

nums = [3, 2, 6, 8, 4, 6, 2]

evens = list(filter(is_even, nums))
double = list(map(lambda n: n*2, evens))

sums = reduce(lambda a, b: a+b, double)

sums

print(evens)
print(double)
print(sums)
```

## Python Decorators

```python
def div(a, b):
    print(a/b)
div(4, 2)
```

```python
def div(a, b):
    print(a / b)
div(2, 4)
```

* Logic apply here should be `numerator > denomenator`
* Means in the above function when i passed div(4, 2) = 2 and it should also give div(2, 4) = 2

```python
def div(a, b):
    if a < b:
        a, b = b, a
    print(a / b)

div(2, 4)
```

```python
# using help  of the decorator you can add the extra feature in the existing function

def div(a, b):
    print(a / b)

def div_decorator(func):
    def inner(a, b):
        if a<b:
            a, b = b,a
        return func(a, b)
    return inner

div = div_decorator(div)
div(2, 4)
```

```python
def my_decorator(func):
    def wrapper():
        print("something before")
        func()
        print("something after")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
```
