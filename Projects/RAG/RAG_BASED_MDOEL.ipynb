```python
import os

from langchain_community.document_loaders import TextLoader, PyPDFLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from langchain_groq import ChatGroq
from langchain_classic.chains import RetrievalQA
```

```python
os.environ["GROQ_API_KEY"] = "YOUR groq api key"
```

```python
# Fetch the PDF from the URL
import requests
url = "https://www.cse.iitd.ac.in/~pkalra/csl101/Python-OOP.pdf"
response = requests.get(url)
```

```python
# Save the PDF to a local file
with open("Python_inbuild_function.pdf", "wb") as f:
  f.write(response.content)
```

```python
# Loading the document

loader = PyPDFLoader("Python_inbuild_function.pdf")

documents = loader.load()

print(documents)
```

```python
text_splitter = CharacterTextSplitter(
    chunk_size=2000,
    chunk_overlap=400
)
```

```python
texts = text_splitter.split_documents(documents)
```

```python
type(texts)
```

```python
len(texts)
```

```python
texts[4]
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
persist_directory = "vector.db"
```

```python
vectordb = Chroma.from_documents(
    documents=texts,
    embedding=embeddings,
    persist_directory=persist_directory
)
```

```python
#retriever
retriver=vectordb.as_retriever()

#llm from groq
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0
)
```

```python
qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriver,
    return_source_documents=True
)
```

```python
query = "Instance of classes"
response = qa_chain.invoke({"query": query})
print(response["result"])
```
