```python
import matplotlib.pyplot as plt #visualization
import seaborn as sns # distribution visualization

%matplotlib inline
plt.rcParams['figure.figsize'] = 6,2 #Determines the figure size

import warnings
warnings.filterwarnings('ignore') #Filter out warnings so no unnecessary warnign should come
```

```python
stats["InternetUsers"]
```

```python
vis1 = sns.distplot(stats["InternetUsers"], bins=30)
plt.show()
```

```python
vis4 = sns.boxplot(data = stats, x="IncomeGroup", y="BirthRate") # BI-VARIATE Analysis
plt.show() # that circle come differently is outliers
```

```python
plt.figure(figsize=(4,4))
vis5 = sns.lmplot(data=stats, x = "InternetUsers", y = "BirthRate") # The line is called regression line
plt.show()
```

```python
vis5 = sns.lmplot(data=stats, x = "InternetUsers", y = "BirthRate", fit_reg=False)
plt.show()
```

```python
vis5 = sns.lmplot(data=stats, x="InternetUsers", y="BirthRate",fit_reg=True, hue="IncomeGroup")
plt.show()
```
