# Amazon Product Sale Analysis 

# Understanding Attribute of the dataset

* product_id - Product ID
* product_name - Name of the Product
* category - Category of the Product
* discounted_price - Discounted Price of the Product
* actual_price - Actual Price of the Product
* discount_percentage - Percentage of Discount for the Product
* rating - Rating of the Product
* rating_count - Number of people who voted for the Amazon rating
* about_product - Description about the Product
* user_id - ID of the user who wrote review for the Product
* user_name - Name of the user who wrote review for the Product
* review_id - ID of the user review
* review_title - Short review
* review_content - Long review
* img_link - Image Link of the Product
* product_link - Official Website Link of the Product

## Bivariate Analysis between `discounted_price_rupee` vs `actual_price_rupee`

```python
amazon_df_new["actual_price_rupee"].mean()
```

```python
amazon_df_new[["discounted_price_rupee", "actual_price_rupee"]].corr()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax = sns.scatterplot(data=amazon_df_new, x="discounted_price_rupee", y="actual_price_rupee")
```

```python
amazon_df_new["price_behavior"] = "Normal"

amazon_df_new.loc[
    (amazon_df_new["discounted_price_rupee"] < 200) &
    (amazon_df_new["actual_price_rupee"] > 20000),
    "price_behavior"
] = "Huge_discount"

amazon_df_new.loc[
    (amazon_df_new["discounted_price_rupee"] >= 300) &
    (amazon_df_new["actual_price_rupee"] <= 2000),
    "price_behavior"
] = "Suspicious"
```

```python
amazon_df_new.head(2)
```

```python

fig = sns.jointplot(data=amazon_df_new,
                    x="discounted_price_rupee",
                    y="actual_price_rupee",
                    kind="scatter", 
                    hue="price_behavior",
                    palette={
                        'Huge_discount': 'blue',
                        'Suspicious': 'red',
                        'Normal': 'gray'
                    },
                    s=70,
                    alpha=0.75,
                    height=8
                    )
plt.xticks(range(0, 1000, 100))
plt.yticks(range(0, 140000, 10000))
plt.title("Colored Price Behavior Scatter Plot")
plt.xlabel("Discounted Price (₹)")
plt.ylabel("Actual Price (₹)")
plt.legend(title="Behavior Type")
plt.show();
```

<ul>
    <li style="color: blue;">Huge_discount</li>
    <ul>
        <li>Initially product priced was very high but discounted really low.</li>
    </ul>
    <li style="color: red">Suspicious</li>
    <ul>
        <li>Intially product priced was low but also discounted highly</li>
    </ul>
</ul>

## Bivariate Analysis between `rating` and `rating_count`

```python
amazon_df_new[['rating', 'rating_count']].corr()
```

```python
ax = sns.jointplot(data=amazon_df_new, x="rating", y="rating_count", kind="scatter", height=8)
plt.yscale('log');
```

The scatter plot between rating and rating_count shows a weak positive correlation.

Products with higher ratings tend to have more reviews, but the relationship is not strong due to the presence of many outliers and a wide variation in rating_count.

Some products with low ratings still have thousands of reviews, while some highly rated items have very few reviews. This indicates that rating_count is influenced by product popularity and time on market, not just rating.
